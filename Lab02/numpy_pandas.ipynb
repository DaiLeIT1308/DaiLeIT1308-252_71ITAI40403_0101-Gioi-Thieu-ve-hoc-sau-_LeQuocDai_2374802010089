# BTVN 1
Cho ma trận = [[99,99,99],
                
                [99,99,99],
                
                [99,99,99]]
Hãy tưởng tượng ô 99 trống 

Giả sử 1 là x

Giả sử 0 là o

Lấy đầu vào từ phía x và phía o luân phiên

Nếu đầu vào từ phía x là (0,0)

Ma trận sẽ trở thành

[[99,99,99],

[99,99,99],

[99,99,99]]

Nếu đầu vào từ phía o cũng là (0,0)

Thông báo họ nhập lại

Nếu không thì điền vào ma trận

Thử thách: Nếu người đầu tiên có 3 phần tử liên tiếp thì dừng trò chơi.

```python
import numpy as np

# Khởi tạo bàn cờ
board = np.full((3,3), 99)

def check_win(board, player):
    # Kiểm tra hàng
    for i in range(3):
        if np.all(board[i, :] == player):
            return True
    
    # Kiểm tra cột
    for i in range(3):
        if np.all(board[:, i] == player):
            return True
    
    # Kiểm tra đường chéo
    if np.all(np.diag(board) == player):
        return True
    if np.all(np.diag(np.fliplr(board)) == player):
        return True
    
    return False

player = 1  # X đi trước

while True:
    print(board)
    row = int(input("Nhập hàng (0-2): "))
    col = int(input("Nhập cột (0-2): "))

    # Kiểm tra ô trống
    if board[row, col] != 99:
        print("Ô đã được đánh, nhập lại!")
        continue

    board[row, col] = player

    # Kiểm tra thắng
    if check_win(board, player):
        print(board)
        print("Người chơi", "X" if player == 1 else "O", "thắng!")
        break

    # Đổi lượt
    player = 0 if player == 1 else 1
```

# BTVN 4 
Tạo một mảng numpy có kích thước 150x5. Giả sử mã này chứa 150 mẫu dữ liệu của sinh viên bao gồm chiều cao, cân nặng, tuổi, lương và điểm trung bình

Tách mảng dựa trên 4 cột đầu tiên thành một biến có tên là x và cột cuối cùng thành y


Tách x thành x_train và x_test trong đó x_train chứa 70% dữ liệu và tách y thành y_trend và y_test, trong đó y_train chứa 70% dữ liệu


Tạo 10 tệp dữ liệu x_train không chồng chéo nhau 

```python
import numpy as np

# Đặt seed để kết quả có thể lặp lại
np.random.seed(42)

# Tạo dữ liệu giả cho 150 sinh viên
# [chiều cao, cân nặng, tuổi, lương, điểm trung bình]
data = np.random.rand(150, 5)

data.shape
```

```python
# 4 cột đầu là X
X = data[:, :4]

# Cột cuối là y
y = data[:, 4]

X.shape, y.shape
```

```python
# Số mẫu train
train_size = int(0.7 * len(X))  # 105 mẫu

# Trộn dữ liệu trước khi chia
indices = np.random.permutation(len(X))

X = X[indices]
y = y[indices]

# Chia dữ liệu
X_train = X[:train_size]
X_test  = X[train_size:]

y_train = y[:train_size]
y_test  = y[train_size:]

X_train.shape, X_test.shape
```

```python
# Chia X_train thành 10 tập con không chồng chéo
x_train_splits = np.array_split(X_train, 10)

# Kiểm tra kích thước từng tập
for i, split in enumerate(x_train_splits):
    print(f"Tập {i+1}: shape = {split.shape}")
```
